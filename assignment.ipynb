```python
import sys
import pyodbc
import datetime

class Hotel:

    def __init__(self):
        print("Welcome to Hotel ABC")

    def DisplayMenu(self):
        try:
            self.conn = pyodbc.connect(
                'Driver={ODBC Driver 17 for SQL Server};'
                'Server=DESKTOP-HO3N392\\SQLEXPRESS;'
                'Database=hotel_python;'
                'Trusted_Connection=yes;'
            )

            self.cursor = self.conn.cursor()
            self.cursor.execute("SELECT * FROM menu")
            self.menu = self.cursor.fetchall()

            print("\n------ MENU ------")
            for index, item in enumerate(self.menu, 1):
                print(f"{index}. {item[1]} - Rs.{item[2]}")

        except Exception as e:
            print("Error in DisplayMenu:", e)


    def TakeOrder(self):
        self.orders = []

        while True:
            try:
                item_index = int(input("\nEnter item number: "))
                quantity = int(input("Enter quantity: "))

                if item_index < 1 or item_index > len(self.menu):
                    print("Invalid item number!")
                    continue

                selected_item = self.menu[item_index - 1]
                menu_id = selected_item[0]
                price = selected_item[2]
                total_price = price * quantity
                self.orders.append((menu_id, quantity, total_price))

                more = input("Do you want anything else? (yes/no): ").lower()
                if more == "no":
                    break
            except Exception as e:
                print("Error in TakeOrder:", e)
        print(self.orders)

    def InsertIntoOrders(self):
        try:
            for order in self.orders:
                menu_id, quantity, total_price = order

                self.cursor.execute("""
                    INSERT INTO orders (menu_id, quantity, price, created_date)
                    VALUES (?, ?, ?, ?)
                """, menu_id, quantity, total_price, datetime.date.today())
 
            self.conn.commit()
            print("\n Order inserted successfully!")

        except Exception as e:
            print("Error inserting order:", e)

        finally:
            self.cursor.close()
            self.conn.close()

obj = Hotel()
obj.DisplayMenu()
obj.TakeOrder()
# obj.InsertIntoOrders()
```

```python
class BillingSystem:
    def __init__(self):
        # Data initialization (Attributes)
        self.subject_book_charge = {
            'Hindi':   {'1-4': 60,  '5-8': 100, '9-10': 150},
            'Marathi': {'1-4': 60,  '5-8': 100, '9-10': 150},
            'English': {'1-4': 80,  '5-8': 100, '9-10': 150},
            'Science': {'1-4': 90,  '5-8': 120, '9-10': 200},
            'Maths':   {'1-4': 100, '5-8': 140, '9-10': 250}
        }
        
        self.notebook_charge = {
            'square':       {100: 40, 200: 70},
            '4lines':       {100: 30, 200: 50},
            '2lines':       {100: 30, 200: 50},
            'single lines': {100: 60, 200: 100},
            'A4 notebook':  {100: 100, 200: 180}
        }
        
        # Totals
        self.total_books_price = 0
        self.total_notebooks_price = 0
        self.bill_details_books = {}
        self.bill_details_notebooks = {}

    def get_user_input(self):
        """Handles the initial input from the user."""
        self.standard = input("Which standard (1-4, 5-8, 9-10): ")
        self.pages = int(input("Notebook type (100 or 200 pages): "))

    def calculate_books(self):
        """Processes the quantities and costs for subject books."""
        print("\n--- Subject Books Section ---")
        for subject in self.subject_book_charge:
            qty = int(input(f"Quantity for {subject} books: ") or 0)
            price_per_unit = self.subject_book_charge[subject][self.standard]
            
            self.total_books_price += (price_per_unit * qty)
            if qty > 0:
                self.bill_details_books[subject] = {"Qty": qty, "PriceEach": price_per_unit}

    def calculate_notebooks(self):
        """Processes the quantities and costs for notebooks."""
        print("\n--- Notebooks Section ---")
        for n_type in self.notebook_charge:
            qty = int(input(f"Quantity for {n_type} notebooks: ") or 0)
            price_per_unit = self.notebook_charge[n_type][self.pages]
            
            self.total_notebooks_price += (price_per_unit * qty)
            if qty > 0:
                self.bill_details_notebooks[n_type] = {"Qty": qty, "PriceEach": price_per_unit}

    def display_bill(self):
        print("\n" + "="*30)
        print("SUMMARY")
        print(f"Books: {self.bill_details_books}")
        print(f"Total Books Cost: {self.total_books_price}")
        print(f"Notebooks: {self.bill_details_notebooks}")
        print(f"Total Notebooks Cost: {self.total_notebooks_price}")
        print("-" * 30)
        final_total = self.total_books_price + self.total_notebooks_price
        print(f"YOUR FINAL BILL IS: {final_total}")
        print("="*30)

# --- Execution ---
if __name__ == "__main__":
    store = BillingSystem()
    store.get_user_input()
    store.calculate_books()
    store.calculate_notebooks()
    store.display_bill()
```
